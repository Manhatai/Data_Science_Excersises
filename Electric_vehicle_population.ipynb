```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
sns.set()
```

```python
df_raw = pd.read_csv('electric_vehicle_population_data.csv')
df_raw
```

```python
df = df_raw.copy()
df
```

```python
df.columns = [col.lower().replace(' ', '_') for col in df.columns]
df
```

```python
df.info()
```

```python
df.columns
```

```python
df = df.rename(columns={"vin_(1-10)": "vin", 'clean_alternative_fuel_vehicle_(cafv)_eligibility': 'cafv_eligibility'})
df
```

```python
df = df.set_index('vin')
df
```

```python
df = df.drop(columns=['electric_utility', '2020_census_tract', 'vehicle_location', 'legislative_district'])
df
```

```python
df = df[df.electric_range != 0]
df
```

```python
df.info()
```

```python
df.dropna()
df
```

```python
df
```

```python
df.cafv_eligibility.value_counts()
```

```python
df.loc[:, 'cafv_eligibility'] = df.cafv_eligibility.replace({'Clean Alternative Fuel Vehicle Eligible': 'Y', 'Not eligible due to low battery range': 'N'})
df
```

```python
df.base_msrp.value_counts()
```

```python
df = df.drop(columns=(['base_msrp', 'dol_vehicle_id', 'postal_code', 'county', 'city']))
```

```python
df.state.value_counts().head(10)
```

```python
df.state.value_counts().plot(kind='pie', title='Wykres ilości zarejestrowanych samochodów w poszczególnych stanach') # Wykres ten pokazuje, że ponad 99.9% samochodów elektrycznych zarejestrowanych jest w Waszyngtonie (WA).
```

```python
df
```

```python
df_numeric = df.copy()

df_numeric['make'] = df_numeric['make'].astype('category').cat.codes # .cat.codes konwertuje dane ze stringa na integer, poprzez przypisanie do poszczególnych danych wartości 0, 1, 2 itd.
df_numeric['model'] = df_numeric['model'].astype('category').cat.codes
df_numeric['electric_vehicle_type'] = df_numeric['electric_vehicle_type'].astype('category').cat.codes
df_numeric['state'] = df_numeric['state'].astype('category').cat.codes
df_numeric['cafv_eligibility'] = df_numeric['cafv_eligibility'].astype('category').cat.codes

df_numeric = df_numeric.rename(columns={'make': 'car_brand'})

df_numeric
```

```python
df_numeric.corr()
```

```python
plt.figure(figsize=(15, 8)) 
sns.heatmap(df_numeric.corr(), annot=True, cmap="YlGnBu") # Poniższa heatmapa pokazuje ciekawe korelacje, praktyczny brak korelacji między typem samochodu i statusem "Clean Alternative Fuel Vehicle" a duży między pojemnością baterii a statusem "Clean Alternative Fuel Vehicle", brak korelacji między typem samochodu a pojemnością baterii elektrycznej, zauważalną korelację między pojemnością baterii a marką samochodu.
```

```python
df_car = df[['car_brand', 'electric_range']]
df_car
```

```python
df_car.electric_range.value_counts()
```

```python
df_agg = df_car.groupby('car_brand').agg({'electric_range': ['sum', 'count']})
df_agg.columns = ['total_electric_range', 'count']
df_agg
```

```python
df_car.columns
```

```python
df_agg['Mean'] = (df_agg['total_electric_range'] / df_agg['count'])
df_agg.head()
```

```python
df_agg['Mean'].plot(kind='bar', figsize=(20, 12), title='Średnia pojemność silników elektrycznych w zależności od marki (w milach)')
```

```python
df['cafv_eligibility'] = df['cafv_eligibility'].replace({'Y': 1, 'N': 0})
df['cafv_eligibility'] =  df['cafv_eligibility'].astype(bool)
df
```

```python
px.bar(data_frame=df_brand_counts, x='car_brand', y='count', title="Ilość samochodów elektrycznych każdej marki w USA posiadających status 'Clean Alternative Fuel Vehicle'", color='count')
```
